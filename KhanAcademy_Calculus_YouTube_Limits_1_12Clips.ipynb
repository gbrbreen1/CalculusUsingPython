# Differential Calculus

Differential calculus is all abour instantanious change التغير اللحظي الآن  ===> what is happening in this instance average speed/ average slope (متوسط السرعة على المسافة كلها) is different than instantanious speed (السرعة اللحظية في لحظة ما)

where the changes of x  and y get smaller and smaller and smaller till it reaches 0. So, we want to take the limit as delta x approaches 0  (not equal to 0) as we cannot divide by 0...... In this time, we calculate the instantanious slope (the derivative)

The derivative is the slope around the point as we take the limit of points closer and closer to the point at question

```python
# the limit means : as x approaches a number (e.g.0), what does y approach (as the limit of x is 0 what is the limit of y) knowing
# that if we apply the function on x=0, we will not get the same result of the function 
# (e.g. f(x)=x**2 when x doesn't equal 0 & f(x)=2 when x equals 0)
from sympy import *
x=Symbol('x')
f=Symbol('f')
init_printing()
f=(x-1)/(x-1)
print(f.subs(x,0))
plot(f, (x,0,10),ylabel='f(x)', ylim=(-5, 5)) # So, we plot the function using the numbers from 0 to 10 and limit the extent of y axis -5,5
limit(f,x,0) # the limit of the function as x approaches 0
```

```python
f=x**2
plot(f, (x,0,10),ylabel='f(x)')
print(f.subs(x,0))
limit(f,x,0)
```

```python
f=1/x
plot(f, (x,-10,10),ylabel='f(x)', ylim=(-10, 10)) # we limit the y so we can see the graph better but if we leave it like before, the details will not be clear
print(f.subs(x,0))
print(limit(f,x,0))
limit(f,x,0, dir='+')# using plus means in the plus direction only
limit(f,x,0, dir='+-') # using plus means in the plus direction and minus means in the minus direction
```

```python
f=1/x**2
plot(f, (x,-10,10),ylabel='f(x)', ylim=(-10, 10)) 
print(f.subs(x,0)) # note that ZOO means complex infinity
limit(f,x,0)
```

```python
f=(x**2-6*x+9)/(x**2-9)
plot(f, (x,-10,10),ylabel='f(x)', ylim=(-10, 10)) 
print(f.subs(x,3))
limit(f,x,3)
```

```python
f=(x**2+x-2)/(x-1)
plot(f, (x,-10,10),ylabel='f(x)', ylim=(-10, 10)) 
print(f.subs(x,1)) # nan means not a number. Most indeterminate forms, such as 0/0 or oo - oo` produce NaN.  Two exceptions 
# are ``0**0 and oo**0, which all produce 1 
limit(f,x,1)
```

```python
from sympy import oo
f=(x**2+3)/x**3
plot(f, (x,-10,10),ylabel='f(x)', ylim=(-10, 10)) 
print(f.subs(x,oo))
limit(f,x,oo)
```

```python
f=sin(x)/x
plot(f, (x,-10,10),ylabel='f(x)', ylim=(-1, 1)) 
print(f.subs(x,0)) 
limit(f,x,0)
```

```python
f=(x-2*abs(x))/abs(x)
plot(f, (x,-10,10),ylabel='f(x)', ylim=(-5, 5)) 
print(f.subs(x,0)) 
print(limit(f,x,0, dir='+')) # when approaching from the right side , the limit is -1
print(limit(f,x,0, dir='-')) # when approaching from the left side, the limit is -3
```

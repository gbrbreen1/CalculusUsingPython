<h3>2-3 Data Standardization/ Data Transformation</h3>


We want to transform mpg to L/100km by mathematical operation because the fuel consumption is measured with L/100km (not mpg) in Egypt. The formula for unit conversion is L/100km = 235 / mpg. So, we need to transform the columns of "city-mpg", "highway-mpg".

```python
# checking the current header of the dataframe before the transformation
df.head()
```

```python
# data transformation and adding column 'city-L/100km' to the dataframe
df['city-L/100km'] = 235/df["city-mpg"]
df["highway-mpg"] = 235/df["highway-mpg"]

# rename column name from "highway-mpg" to "highway-L/100km"
df.rename(columns={'highway-mpg':'highway-L/100km'}, inplace=True)

df.head()
```

<h3>2-4 Data Normalization </h3>

We want to scale/ normalize the columns "length", "width" and "height" so their value ranges from 0 to 1. To do this, we will replace the origianl value by (original value)/(maximum value)

```python
# normalizing "length", "width", and "height"
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max() 
# show the scaled columns
df[["length","width","height"]].head()
```

<h3>2-5 Data Binning</h3>

The client want to bin (narrow the intervals from 57 of) the "horsepower" to 3 categories (high horsepower, medium horsepower, and little horsepower) as he just wants to check the price based ONLY on these 3 categories (not based on the real values of the column).

**2-5-1 Binning the column (narrawing its values to a specific number of categories) from 57 to 3:**

```python
# checking the data type of the column
print("The data type of horsepower is\n")
df["horsepower"].dtype
```

```python
binwidth = (max(df["horsepower"])-min(df["horsepower"]))/4

# building a bin (the end of a category and the end of the next) array, with a minimum value to a maximum value, with binwidth calculated above
bins = np.arange(min(df["horsepower"]), max(df["horsepower"]), binwidth)
bins
```

**P-value and the significance of the correlation**

By convention, when the

- p-value is < 0.001 we say there is strong evidence that the correlation is 
significant,
- the p-value is < 0.05; there is moderate evidence that the correlation is significant,
- the p-value is < 0.1; there is weak evidence that the correlation is significant, and
- the p-value is > 0.1; there is no evidence that the correlation is significant.

We will use the below criteria to express the correlation strength:


1.   correlation < 0.3 ==> None or very weak
2.   0.3 < correlation < 0.5  ==> Weak
3.   0.5 < correlation < 0.7 ==> Moderate
4.   correlation > 0.7 ==> Strong

```python
df.dtypes
```

below we will calculate Pearson Correlation Coefficient and P-value of every numeric feature and "price"

```python
from scipy import stats

# calculate the Pearson Correlation Coefficient and P-value of 'wheel-base' and 'price'
pearson_coef, p_value = stats.pearsonr(df['wheel-base'], df['price'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)  
```

##### Conclusion: 
Since the p-value (8.076488270733218*10^-20) is < 0.001, the correlation between wheel-base and price is statistically significant, although the linear relationship isn't extremely strong (~0.585) as the correlation is moderate (<0.7)

```python
#calculate the Pearson Correlation Coefficient and P-value of 'horsepower' and 'price'
pearson_coef, p_value = stats.pearsonr(df['horsepower'], df['price'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)  
```

<h3>3-5 Analysis of Variance (ANOVA) for ctaegorical/ nominal/ ordinal variables<h3>




We will calculate ANOVA to test whether there are significant differences between the means of two or more groups. ANOVA assumes that the means of all groups are the same, calculates how much the actual means deviate from the assumption, and reports it as the **F-test score**. A larger score means there is a larger difference between the means.

We will also calculate the **P-value** as it tells how statistically significant is our calculated score value (F-test score).

**Note**: If our price variable (the dependent variable) is strongly correlated with the independent categorical variable we are analyzing, expect ANOVA to return a sizeable F-test score and a small p-value.

```python
# group a partial dataframe by the variable "drive-wheels" and calculate the average price
df_gptest=df[['drive-wheels','body-style','price']]
grouped_test_DW=df_gptest.groupby(['drive-wheels'])
grouped_test_DW.head(2)
```

```python
# Construct DataFrame from group with provided name (4wd)
grouped_test_DW.get_group('4wd')['price']
```

```python
# calculate the One-way ANOVA (F-test score & P-value) for ALL the groups by creating a dataframe for each of them (on the fly)
f_val, p_val = stats.f_oneway(grouped_test_DW.get_group('fwd')['price'], grouped_test_DW.get_group('rwd')['price'], grouped_test_DW.get_group('4wd')['price'])  

print( "ANOVA results: F=", f_val, ", P =", p_val)   
```

This is a great result, with a large F test score showing a strong correlation and a P value of almost 0 implying almost certain statistical significance. But does this mean all three tested groups are all this highly correlated?

```python
# calculate the OnF-te-way ANOVA (est score & P-value) for fwd and rwd groups
f_val, p_val = stats.f_oneway(grouped_test_DW.get_group('fwd')['price'], grouped_test_DW.get_group('rwd')['price'])  

print( "ANOVA results: F=", f_val, ", P =", p_val)   
```

```python
# calculate the OnF-te-way ANOVA (est score & P-value) for 4wd and rwd
f_val, p_val = stats.f_oneway(grouped_test_DW.get_group('rwd')['price'], grouped_test_DW.get_group('4wd')['price'])  

print( "ANOVA results: F=", f_val, ", P =", p_val)   
```

```python
# calculate the Oneway ANOVA (F-test score & P-value) for 4wd and fwd
f_val, p_val = stats.f_oneway(grouped_test_DW.get_group('fwd')['price'], grouped_test_DW.get_group('4wd')['price'])  

print( "ANOVA results: F=", f_val, ", P =", p_val)   
```

**we can say that there is a strong correlation between a categorical variable and other variables, if the ANOVA test gives us a large F-test value and a small p-value.**

```python
# creating a function to plot the data
def PlotPolly(model,independent_variable,dependent_variabble, Name):
    x_new = np.linspace(5, 20, 100)
    y_new = model(x_new)

    plt.plot(independent_variable,dependent_variabble,'.', x_new, y_new, '-')
    plt.title('Polynomial Fit with Matplotlib for Price ~ Length')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of Cars')
    # mySize=plt.figure(figsize=(16, 6))

    plt.show()
    plt.close()
    
print("done")
```

 We select the value of Alfa that minimizes the test error (the value that has the maximum R^2), for example, we can use a for loop. 

```python
# create different R^2 values for different alpha/ regularization values
Rsqu_test=[]
Rsqu_train=[]
# dummy1=[]
ALFA=5000*np.array(range(0,10000))
for alfa in ALFA:
    RidgeModel=Ridge(alpha=alfa) 
    RidgeModel.fit(x_train_pr,y_train)
    Rsqu_test.append(RidgeModel.score(x_test_pr,y_test))
    Rsqu_train.append(RidgeModel.score(x_train_pr,y_train))
```

```python
width = 12
height = 10
plt.figure(figsize=(width, height))

plt.plot(ALFA,Rsqu_test,label='validation data  ')
plt.plot(ALFA,Rsqu_train,'r',label='training Data ')
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.legend()
```

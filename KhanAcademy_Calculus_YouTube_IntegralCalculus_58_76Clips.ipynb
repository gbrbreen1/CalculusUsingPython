# Indefinite integrals

We use integral when we have the derivative of a function and we want to get the function. Integral calculus is the opposite of differential calculus. 

Evaluating indefinite integrals  doesn’t give you a number (like in the definite). Instead, when you evaluate an indefinite integral, the result is a function that you can use to obtain all related definite integrals. Indefinite integrals provide a convenient way to calculate definite integrals. In fact, the indefinite integral is the inverse of the derivative

```python
from sympy import *
x=Symbol('x')
fDash=Symbol('fDash')
init_printing()
fDash=x**2+x+1

plot(fDash,x, (x,-10,10),ylabel='f(x)', ylim=(-100, 100))
integrate(fDash,x)
```

```python
fDash=(log(2))**2/x
plot(fDash,x, (x,-10,10),ylabel='f(x)', ylim=(-10, 10))
integrate(fDash,x)
```

```python
fDash=(2*x+3)*(x**2+3*x+15)**5
plot(fDash,x, (x,-10,10),ylabel='f(x)', ylim=(-10, 10))
simplify(integrate(fDash,x))
```

```python
fDash=sin(x)/(cos(x))**2
plot(fDash,x, (x,-50,50),ylabel='f(x)', ylim=(-10000, 10000))
integrate(fDash,x)
```

# Definite integral

A definite integral has upper and lower limits on the integrals, and it’s called definite because, at the end of the problem, we have a number – it is a definite answer.

The definite integral of f(x) is a NUMBER and represents the area under the curve f(x) from x=a to x=b.

Again, the definite integral evaluates to a number, which is the area under the function between the limits of integration.

The definite integral represents the area on a graph bounded by a function, the x-axis, and two vertical lines called the limits of integration.

The area problem is all about finding the area under a continuous function between two constant values of x that are called the limits of integration, usually denoted by a and b.

A definite integral definitely equals a number. at the end of the day, a number is just a number. And, because a definite integral is a measurement of area, you should expect the answer to be a number.

```python
from sympy import *
x=Symbol('x')
fDash=Symbol('fDash')
init_printing()
fDash=x**2+x+1

plot(fDash,(x,-5,5), (x,-5,5),ylabel='f(x)', ylim=(-100, 100))
float(integrate(fDash,(x,-5,5)))# this is the area under the curve between -5,5 on x axis
```

# Calculate area under the curve for definite integral

```python
from sympy import *
x=Symbol('x')
fDash=Symbol('fDash')
init_printing()
fDash=-(16-x**2)**0.5

plot(fDash,(x,-4,4), (x,-4,4), ylabel='f(x)', ylim=(-5, 5))
float(integrate(fDash,(x,-4,4)))# this is the area under the curve between -4,4 on x axis

# notice that So, area above the x-axis is positive, but area below the x-axis is measured as negative area
# The definite integral takes this important distinction into account. It provides not just the area but the signed area of 
# a region on the graph
```

```python
from sympy import *
x=Symbol('x')
fDash=Symbol('fDash')
init_printing()
fDash=(16-x**2)**0.5

plot(fDash,(x,-4,4), (x,-4,4), ylabel='f(x)', ylim=(-5, 5))
float(integrate(fDash,(x,-4,4)))# this is the area under the curve between -4,4 on x axis
```

Solve the definite integral as an indefinite integral evaluated between the given limits of integration ==> the best more accurate method to calculate the area under the curve

What is written in books that this is the best way but in sympy, it is very slow (note that we get the same result but faster using the above way by calculating the area using definite  integral)

```python
from sympy import *
x=Symbol('x')
fDash=Symbol('fDash')
init_printing()
fDash=(16-x**2)**0.5

plot(fDash,(x,-4,4), (x,-4,4), ylabel='f(x)', ylim=(-5, 5))

# get the indefinite integral
print(integrate(fDash,x))# this is the area under the curve in general using indefinite integral

# evaluate the indefinite integral you got for the limits of integration
print(integrate(fDash,x).subs(x,4))
print(integrate(fDash,x).subs(x,-4))

# Plug these limits of integration into the expression and simplify it
myExpr=integrate(fDash,x).subs(x,4)-integrate(fDash,x).subs(x,-4)
print(simplify(myExpr))
print(float(simplify(myExpr))) # note that it is the same result as above but it is sower using sympy
```

# Finding the area between 2 functions

```python
from sympy import *
x=Symbol('x')
fDash1=Symbol('fDash1')
fDash2=Symbol('fDash2')
init_printing()
fDash1=x**2
fDash2=x**0.5
fDash1Plot=plot(fDash1,(x,-4,4), (x,-4,4), ylabel='f(x)', ylim=(-5, 5))
fDash2Plot=plot(fDash2,(x,-4,4), (x,-4,4), ylabel='f(x)', ylim=(-5, 5))

fDash2Plot[0].line_color = 'red'
fDash1Plot.append(fDash2Plot[0])

# check if the functions intersect
fDash1Plot.show() # as shown, there is an area between the two curves and we need to calculate this area

```

# Calculating the length of a part of a function (Arc length)

```python
from sympy import *
x=Symbol('x')
fDash=Symbol('fDash')
init_printing()
fDash=x**2

plot(fDash,(x,-4,4), (x,-4,4), ylabel='f(x)', ylim=(-5, 5))

print(float(Curve((fDash, x), (x, 0, 2)).length)) # this the length of the curve between x=0 and x=2
```

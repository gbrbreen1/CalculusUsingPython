# Maxima, minima, and saddle

```python
# finding the maxima, minima, and saddle
# https://www.mathsisfun.com/calculus/maxima-minima.html

# In a smoothly changing function a maximum or minimum is always where the function flattens out  (except for a saddle point).
# In other words, when its derivative equal 0        
# use this function: f(x) = 3 + 14x  5x**2
```

```python
from sympy import *
x=Symbol('x')
f=Symbol('f')
init_printing()
f=3 + 14*x-5*x**2
print(diff(f,x))
print(solve(diff(f,x),x))
# first, we get the first derivative then we solve the equation for x when y=0 ==>the point where the derivative is equal to 0
float(solve(diff(f,x),x)[0]) #===> x=1.4
```

```python
# second, we find the second derivative f''(x)
# When a function's slope is zero at x, and the second derivative at x is (we evaluate the second derivative for x and see the result):
    #less than 0, it is a local maximum
    #greater than 0, it is a local minimum
    #equal to 0, then it is a saddle point and the test fails (there may be other ways of finding out though)
```

```python
# evaluate the second derivative at x=1.4
#print(diff(diff(f,x),x))
print(diff(diff(f,x),x).subs(x,1.4)) # @ x=1.4, as the result is less than 0 ==> it is a local maximum
```

```python
# Another example for maxima, minima, saddle
from sympy import *
x=Symbol('x')
f=Symbol('f')
init_printing()
f = x**3-6*x**2+12*x-5
print(diff(f,x))
print(float(solve(diff(f,x),x)[0])) # now we have the value of x when y=0 ===> x=2
```

```python
print(diff(diff(f,x),x))# the second derivative
print(diff(diff(f,x),x).subs(x,2))# @ x=2, as the result is 0 ===> it is a saddle point       
```

```python
# http://tutorial.math.lamar.edu/Classes/CalcI/AbsExtrema.aspx
# Finding Absolute Extrema of f(x) on an interval [a,b]:
    # Find all critical points of f(x) that are in the interval [a,b] This makes sense if you think about it. Since we are only 
    #interested in what the function is doing in this interval we dont care about critical points that fall outside the interval.
    # Evaluate the function at the critical points found in step 1 and the end points of the interval
    # Identify the absolute extrema.
```

```python
# let us first know how to plot a function in a specific interval[-51,51]
from sympy import *
x=Symbol('x')
f=Symbol('f')
init_printing()
f=3 + 14*x-5*x**2
myFunction=plot(f,(x,-51,51),ylabel='f(x)', ylim=(-10000, 10000)) # plot of the function
print(solve(diff(f,x),x))
print(float(solve(diff(f,x),x)[0]))
print(diff(diff(f,x),x).subs(x,float(solve(diff(f,x),x)[0]))) # @ x=1.4 as the result is less than 0 ==> it is a maxima
```

```python
# get the maximum of a function using Scipy
# https://stackoverflow.com/questions/10146924/finding-the-maximum-of-a-function
import scipy as sc
import scipy.optimize as opt
from sympy import *
x=Symbol('x')
f=Symbol('f')
init_printing()
f=3 + 14*x-5*x**2
myFunction=plot(f,(x,-51,51),ylabel='f(x)', ylim=(-10000, 10000)) # plot of the function


# this is how to use lambda
# >>> sum = lambda x, y : x + y
# >>> sum(3,4)
# 7
def f(x): return 3 + 14*x-5*x**2
max_x = opt.fmin(lambda x: -f(x), 0)[0] #            
print('The maximum value is:  ',max_x)
# array([ 1.])
```

```python
# get the minimum of a function using Scipy
# https://stackoverflow.com/questions/10146924/finding-the-maximum-of-a-function
import scipy as sc
from sympy import *
x=Symbol('x')
f=Symbol('f')
init_printing()
f=3 + 14*x+5*x**2
myFunction=plot(f,(x,-51,51),ylabel='f(x)', ylim=(-10000, 10000)) # plot of the function

def f(x): return 3 + 14*x+5*x**2
min_x = opt.fmin(lambda x: f(x), 0)[0] #              
print('The minimum value is: ',min_x)
# array([ 1.])
```

```python
# get the maximum and the minimum using sympy                         
# https://www.mathsisfun.com/calculus/maxima-minima.html
from sympy import *
x=Symbol('x')
f=Symbol('f')
init_printing()
f=5*x**3+2*x**2-3*x
plot(f,(x,-10,10),ylabel='f(x)', ylim=(-10, 10))
print(solve(diff(f,x),x))
print(float(solve(diff(f,x),x)[0]))
print(diff(diff(f,x),x).subs(x,float(solve(diff(f,x),x)[0]))) # @ x=-3/5 ===> a maxima
print(float(solve(diff(f,x),x)[1]))
print(diff(diff(f,x),x).subs(x,float(solve(diff(f,x),x)[1])))# @ x=1/3 ===> a minima
```

# finding the equation of a tangent line

```python
# remember that the equation of the line is always 
# y=mx+b
# where m is the slope (or gradient) of hte line = change in y/ change in x
# and b is the y-intercept
```

```python
# https://butterflyofdream.wordpress.com/2014/01/23/sympy-example-finding-tangent-lines-to-a-function/
#  Find an equation for the line that is tangent to the curve at point (-1, 0)
from sympy import *
x=Symbol('x')
y=Symbol('y')
init_printing()

y = x**3 - x
 
# find the derivative of y
dy = diff(y, x)
print('the derivative equation is:  ',dy)
 
# find slope of the tangent line at point (-1,0) and equation 
m = dy.subs(x, -1) # m is the slope of the line in this point where x= -1
print('m:  ',m)

# find the y-intercept 
b=0-(m*-1)
print('b or the y-intercept is: ',b)

# make the equation as you have m and b
#y1 = m*(x + 1)
#print(y1)
print('the equation is: ',m*x+b)
```

# Vector field

Vector field is a function that assigns each point a 2 or 3 dimensional vector that may not make a lot of sense, but most people do know what a vector field is, or at least they’ve seen a sketch of a vector field. If you’ve seen a current sketch giving the direction and magnitude of a flow of a fluid or the direction and magnitude of the winds then you’ve seen a sketch of a vector field.

```python
import numpy as np
import matplotlib.pyplot as plt

n = 8
X, Y = np.mgrid[0:n, 0:n]
T = np.arctan2(Y - n / 2., X - n/2.)
R = 10 + np.sqrt((Y - n / 2.0) ** 2 + (X - n / 2.0) ** 2)
U, V = R * np.cos(T), R * np.sin(T)

plt.axes([0.025, 0.025, 0.95, 0.95])
plt.quiver(X, Y, U, V, R, alpha=.5)
plt.quiver(X, Y, U, V, edgecolor='k', facecolor='None', linewidth=.5)

plt.xlim(-1, n)
plt.xticks(())
plt.ylim(-1, n)
plt.yticks(())

plt.show()
```

# Limits of multivariate functions

We will concentrate mostly on limits of functions of two variables, but the ideas can be extended out to functions with more than two variables

http://tutorial.math.lamar.edu/Classes/CalcIII/Limits.aspx


```python


x=Symbol('x')
y=Symbol('y')
f=Symbol('f')
init_printing()
f=x**2*y**2/(x**4+3*y**4)
plot3d(f, (x, -5, 5), (y, -5, 5))

# let us see what will happen along x axis if we plug y=0 along the axis and see the limit as x approaches 0
print(limit(f.subs(y,0),x,0))

# let us see what will happen along y axis if we plug x=0 along the axis and see the limit as y approaches 0
print(limit(f.subs(x,0),y,0))

# the results are 0 and 0. This does NOT say that the limit exists and has a value of zero. This only means that the limit 
# happens to have the same value along two paths.


# Let’s take a look at a third fairly common path to take a look at. In this case we’ll move in towards the origin along the path
# y=x while x ia approaching 0
print(limit(f.subs(y,x),x,0))

# So, a different value (1/4) from the previous two paths and this means that the limit can’t possibly exist.
```

```python
# finding the partial derivative 
from sympy import *
from sympy.plotting import plot3d
z=Symbol('z')
x= Symbol('x')
y=Symbol('y')
init_printing()

z=x+y
plot3d(z, (x, -5, 5), (y, -5, 5))
# let us find the derivative with respect to x (how the function result changes/ what is the rate of change of the function as 
# we shift around x while y is constant)
print(diff(z,x))

# let us say that x=1 and y=2 ==> we find f(1,2)
diff(z,x).subs(x,1)
```

```python
# let us find the derivative with respect to y (how the fnction result  changes as we shift around y while x is constant)
print(diff(z,y))

# let us say that x=1 and y=2 ==> we find f(1,2)
diff(z,y).subs(y,2)
diff(z,y)
```

# Tangent planes and linear approximation

Here we will talk about the plane that is “tangent” to the surface as a point.

A tangent line to a curve was a line that just touched the curve at that point and was “parallel” to the curve at the point in question. Well tangent planes to a surface are planes that just touch the surface at the point and are “parallel” to the surface at the point. Note that this gives us a point that is on the plane. So, the point is on both the surface and the plane

What we need to do now is determine the equation of the tangent plane.

In Sympy, A plane is a flat, two-dimensional surface. A plane is the two-dimensional analogue of a point (zero-dimensions), a line (one-dimension) and a solid (three-dimensions). A plane can generally be constructed by two types of inputs. They are three non-collinear points and a point and the plane’s normal vector.

# Gradient/ Grad

Be careful that directional derivative of a function is a scalar while gradient is a vector.
For a function u= f(x,y) the partial derivative wrt x gives the rate of change of f in the direction of x. Like the derivative, the gradient represents the slope of the tangent of the graph of the function. More precisely, the gradient points in the direction of the greatest rate of increase of the function, and its magnitude is the slope of the graph in that direction.

The gradient is a way of packing together all the partial derivatives information of a function

So the gradient puts the partial derivative of x and the partial derivative of y, in a function of x and y,  in a vector ===>  nablef(x,y)=vector [partial derivative of x, partial derivative of y]

Note that the vector can be derivatives of x, y , z , ....etc depending on the number of variables (n-dimentional)

Gradient vector is the direction of steepest ascent (to walk uphill of the graph as fast as you can if you are a mountain climber) ==أشد ميل للصعود لأعلى

# Directional derivative/ Directional Gradient

The directional derivative of a multivariate differentiable function along a given vector v at a given point x intuitively represents the instantaneous rate of change of the function, moving through x with a velocity specified by v. So, it is the derivative with respect to this vector ( the x and the y of this vector)
 معدل التغير اللحظي للدالة عند التحرك على المحور السيني بسرعة تتحدد من خلال المتجه المطلوب

Directional derivative for a function in the direction of a vector (a,b) is:
 ===> a* the derivative of the function with respect to x + b* the derivative of the function with respect to y

===> the dot product of the vector and the gradient of the function

while calculating it, if we use the unit vector whose magnitude is 1, we can interpret the directional derivative as the slope in the direction of the vector

Note that the partial derivative is a special case of the directional derivative

Note that the directional derivative is a scalar while the gradient is a vector

https://www.youtube.com/watch?v=4tdyIGIEtNU&list=PLSQl0a2vh4HC5feHa6Rc5c0wbRTx56nF7&index=23

```python
# find the directional derivative of x*cos(y) in the direction of (1,1) at point(-1,-1)
from sympy import *
f=Symbol('f')
x= Symbol('x')
y=Symbol('y')
t=Symbol('t')



f=(x**2)*y
print(f)

import numpy as np
# computing :  a* the partial derivative of the function with respect to x
part1=1*diff(f,x)
print(part1)

# computing: b* the partial derivative of the function with respect to y
part2=1*diff(f,y)
print(part2)

# summing part1 and part2
print(part1+part2)

print('the directional derivative is: ',(part1+part2).subs([(x,-1), (y,-1)])/np.linalg.norm([1,1]))
```

```python
# Another way to find the directional derivative of the function in the direction of the unit vector (1,1) using the gradient at
# point (-1,-1)

from sympy import *
f=Symbol('f')
x= Symbol('x')
y=Symbol('y')
t=Symbol('t')



f=(x**2)*y

import numpy as np
directionVector=Matrix([1,1])
directionVectorArray=np.array(directionVector).astype(np.float64) # convert sympy matrix to numpy array

# computing derivatives with respect to x and y in a vector ===> here we are computing the gradient
partialDerivatives=Matrix([diff(f,x),diff(f,y)])

# computing the gradient at point(-1,-1)
myGradient=partialDerivatives.subs([(x,-1), (y,-1)])
print('the gradient is: ',myGradient)

# computing the magnitude of vector (1,1)
directionVectorMagnitude=np.linalg.norm(directionVectorArray)
print(directionVectorMagnitude)

# computing the directional derivative
print('the directional derivative is: ',directionVector.dot(myGradient)/directionVectorMagnitude)

```
